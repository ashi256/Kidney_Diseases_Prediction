

Attribute Information:


1. Age(numerical) age in years
2. Blood Pressure(numerical) bp in mm/Hg
3. Specific Gravity(nominal) sg - (1.005,1.010,1.015,1.020,1.025)

5. Sugar(nominal) su - (0,1,2,3,4,5)
6. Red Blood Cells(nominal) rbc - (normal,abnormal)
7. Pus Cell (nominal) pc - (normal,abnormal)
8. Pus Cell clumps(nominal) pcc - (present,notpresent)
9. Bacteria(nominal) ba - (present,notpresent)
10. Blood Glucose Random(numerical) bgr in mgs/dl
11. Blood Urea(numerical) bu in mgs/dl
12. Serum Creatinine(numerical) sc in mgs/dl
13. Sodium(numerical) sod in mEq/L
14. Potassium(numerical) pot in mEq/L
15. Hemoglobin(numerical) hemo in gms
16. Packed Cell Volume(numerical)
17. White Blood Cell Count(numerical) wc in cells/cumm
18. Red Blood Cell Count(numerical) rc in millions/cmm
19. Hypertension(nominal) htn - (yes,no)
20. Diabetes Mellitus(nominal) dm - (yes,no)
21. Coronary Artery Disease(nominal) cad - (yes,no)
22. Appetite(nominal) appet - (good,poor)
23. Pedal Edema(nominal) pe - (yes,no)
24. Anemia(nominal) ane - (yes,no)
25. Class (nominal) class - (ckd,notckd)

```python
x = df.drop(['classification'],axis=1)
y = df[['classification']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=.3,random_state = 42)
```

```python
def result(x,y):
  cm = confusion_matrix(x,y)
  tp = cm[0][0]
  fp = cm[0][1]
  fn = cm[1][0]
  tn = cm[1][1]
  accuracy = (tp+tn)/(tp+fp+fn+tn)
  recall = tp/(tp+fn)
  precision = tp/(tp+fp)
  f1_score = (precision*recall)/(precision+recall)
  fpr,tpr,threshold = roc_curve(x,y)
  return accuracy,precision,recall,f1_score,fpr,tpr,threshold
```

Logistic Regression

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)


pred = lr.predict(x_test)


accuracy,precision,recall,f1_Score,Fpr1,Tpr1,threshold1 = result(y_test,pred)
print('Accuracy - {:.2%}  Precision - {:.2%}  Recall - {:.2%}  F1_Score - {:.2%}'.format(accuracy,precision,recall,f1_Score))
```

Feature Importance

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model=ExtraTreesClassifier()
model.fit(x,y)

plt.figure(figsize=(8,6))
ranked_features=pd.Series(model.feature_importances_,index=x.columns)
ranked_features.nlargest(24).plot(kind='barh')
plt.show()
```

Random Forest

```python
rr = RandomForestClassifier()
rr.fit(x_train,y_train)


pred = dr.predict(x_test)


accuracy,precision,recall,f1_Score,Fpr3,Tpr3,threshold3 = result(y_test,pred)
print('Accuracy - {:.2%}  Precision - {:.2%}  Recall - {:.2%}  F1_Score - {:.2%}'.format(accuracy,precision,recall,f1_Score))
```

AdaBoost 

```python
AdaBoost = AdaBoostClassifier()
AdaBoost = AdaBoost.fit(x_train,y_train)


pred = AdaBoost.predict(x_test)


accuracy,precision,recall,f1_Score,Fpr4,Tpr4,threshold4 = result(y_test,pred)
print('Accuracy - {:.2%}  Precision - {:.2%}  Recall - {:.2%}  F1_Score - {:.2%}'.format(accuracy,precision,recall,f1_Score))
```

Gradient Boost

```python
GradientBoost = GradientBoostingClassifier()
GradientBoost = GradientBoost.fit(x_train,y_train)

pred = GradientBoost.predict(x_test)


accuracy,precision,recall,f1_Score,Fpr5,Tpr5,threshold5 = result(y_test,pred)
print('Accuracy - {:.2%}  Precision - {:.2%}  Recall - {:.2%}  F1_Score - {:.2%}'.format(accuracy,precision,recall,f1_Score))
```
